```python
import pandas as pd
meteorites = pd.read_csv('datasets/Meteorite_Landings.csv')
meteorites.tail(5)
```

```python
meteorites.dtypes
```

```python
meteorites = meteorites.join(meteorites['year'].str.split(' ', n=1, expand=True).rename(columns={0:'time_date', 1:'time_exact'}))
```

```python
meteorites.head()
```

```python
meteorites = meteorites.join(meteorites['time_date'].str.split('/', n=2, expand=True).rename(columns={0:'month', 1:'day',2:'year_num'}))
```

```python
meteorites.head()
```

```python
meteorites['year'] = meteorites['year_num'].values
```

```python
meteorites.head()
```

```python
meteorites = meteorites.drop(columns='year_num')
```
