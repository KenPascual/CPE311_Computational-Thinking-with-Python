*   Four people with 20kg of supplies want to cross a river using a boat that can only 100kg.    
*  The people are Roman(90kg), Verlyn(80kg), Lloyd(60kg), and Robin(40kg).

How do they get across?

<br>For instance, Lloyd and Robin can be together in one boat.</br>
<br>Verlyn/Lloyd/Robin and the load of supplies can be together in one boat.</br>
<br>Roman can only go alone.</br>

I thought that in order to lessen the number of round trips made by the boat, the first trip should carry two people so that one can be sent to the other side and the other can return and pick up the others.

Thus, Robin and Lloyd goes first.

```python
import sys

# Define the people and their weights
people = {
    'Roman': 90,
    'Verlyn': 80,
    'Lloyd': 60,
    'Robin': 40
}
load_weight = 20
boat_capacity = 100

# Starting side (left side) and far side (right side)
starting_side = ['Roman', 'Verlyn', 'Lloyd', 'Robin']
far_side = []
load_on_start = True  # Load is initially on the starting side

# Function to print current state
def print_state():
    print("Starting side:", starting_side, "Load" if load_on_start else "")
    print("Far side:", far_side, "Load" if not load_on_start else "")
    print("-" * 40)

# Function to cross the river
def cross_river(persons, carry_load=False):
    total_weight = sum(people[p] for p in persons)
    if carry_load:
        total_weight += load_weight

    # Stop the entire program if the boat is overloaded
    if total_weight > boat_capacity:
        print("Too heavy! Boat will sink. Terminating program.")
        sys.exit(1)  # Exit with error code 1

    # Move people to the far side
    for person in persons:
        if person in starting_side:
            starting_side.remove(person)
            far_side.append(person)

    # Move load if necessary
    global load_on_start
    if carry_load:
        load_on_start = not load_on_start

    print_state()

# Function for a person to return
def return_boat(person):
    if person in far_side:
        far_side.remove(person)
        starting_side.append(person)
        print_state()
    else:
        print(f"{person} is not on the far side! Cannot return.")

# Initial state
print("Initial state:")
print_state()

# Step 1: Robin and Lloyd cross
cross_river(['Robin', 'Lloyd'])

# Robin returns
return_boat('Robin')

# Step 2: Roman crosses alone
cross_river(['Roman'])

# Lloyd returns
return_boat('Lloyd')

# Step 3: Robin and Lloyd cross again
cross_river(['Robin', 'Lloyd'])

# Robin returns
return_boat('Robin')

# Step 4: Verlyn and the load cross
cross_river(['Verlyn'], carry_load=True)

# Lloyd returns
return_boat('Lloyd')

# Step 5: Robin and Lloyd cross
cross_river(['Robin', 'Lloyd'])

# Final state (only prints if the program hasn't exited)
print("Final state:")
print_state()
```

```python
import sys

class Person:
    def __init__(self, name, weight):
        self.name = name
        self.weight = weight

    def __str__(self):
        return self.name

class Boat:
    def __init__(self, capacity):
        self.capacity = capacity
        self.current_load = 0
        self.passengers = []

    def add_passenger(self, person):
        if self.current_load + person.weight <= self.capacity:
            self.passengers.append(person)
            self.current_load += person.weight
            return True
        return False

    def remove_passenger(self, person):
        if person in self.passengers:
            self.passengers.remove(person)
            self.current_load -= person.weight

    def clear_passengers(self):
        """Unloads all passengers from the boat."""
        self.passengers = []
        self.current_load = 0

    def print_boat_status(self):
        print(f"Boat Status: {', '.join(str(p) for p in self.passengers) if self.passengers else 'Empty'} | Load: {self.current_load}/{self.capacity}")

class RiverCrossing:
    def __init__(self):
        # Define people with their weights
        self.roman = Person('Roman', 90)
        self.verlyn = Person('Verlyn', 80)
        self.lloyd = Person('Lloyd', 60)
        self.robin = Person('Robin', 40)
        self.load = Person('Load', 20)

        # Create boat with capacity
        self.boat = Boat(100)

        # Initialize sides of the river
        self.starting_side = [self.roman, self.verlyn, self.lloyd, self.robin, self.load]
        self.far_side = []

        # Crossings and returns represented as sequences of people
        self.crossings = [
            ([self.robin, self.lloyd], self.robin),   # Robin and Lloyd cross, Robin returns
            ([self.roman], self.lloyd),               # Roman crosses, Lloyd returns
            ([self.robin, self.lloyd], self.robin),   # Robin and Lloyd cross again, Robin returns
            ([self.verlyn, self.load], self.lloyd),   # Verlyn and the load cross, Lloyd returns
            ([self.robin, self.lloyd], None)          # Robin and Lloyd cross last time, no return
        ]

    def print_state(self):
        print("Starting side:", [str(person) for person in self.starting_side])
        print("Far side:", [str(person) for person in self.far_side])
        self.boat.print_boat_status()
        print("-" * 40)

    def cross_river(self, people_to_cross):
        self.boat.clear_passengers()  # Unload boat before crossing
        total_weight = sum(person.weight for person in people_to_cross)

        if total_weight > self.boat.capacity:
            print("Too heavy! Boat will sink. Terminating program.")
            sys.exit(1)  # Stop execution

        for person in people_to_cross:
            self.boat.add_passenger(person)
            self.starting_side.remove(person)
            self.far_side.append(person)

        self.print_state()
        self.boat.clear_passengers()  # Unload boat once it reaches the far side

    def return_boat(self, person):
        self.boat.clear_passengers()  # Unload boat before return
        self.starting_side.append(person)
        self.far_side.remove(person)

        self.print_state()
        self.boat.clear_passengers()  # Unload boat after returning

    def start_crossing(self):
        # Initial state
        print("Initial state:")
        self.print_state()

        # Loop through the crossings and returns
        for crossing, returner in self.crossings:
            self.cross_river(crossing)
            if returner:  # If there's a returner, make them return
                self.return_boat(returner)

        # Final step: ensure everyone is on the far side
        self.far_side.extend(self.starting_side)  # All remaining people cross
        self.starting_side.clear()  # Clear the starting side
        print("Final state:")
        self.print_state()

river_crossing = RiverCrossing()
river_crossing.start_crossing()
```
